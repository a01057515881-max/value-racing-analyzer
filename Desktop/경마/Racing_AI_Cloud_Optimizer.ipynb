```python
# 1. 구글 드라이브 연결
from google.colab import drive
import os
drive.mount('/content/drive')

# [설정] 압축 파일이 위치한 드라이브 경로 (수정 필요 시 수정)
DRIVE_PATH = '/content/drive/MyDrive/RacingAI' 
PACKAGE_NAME = 'racing_ai_package.zip'

if not os.path.exists(DRIVE_PATH):
    os.makedirs(DRIVE_PATH, exist_ok=True)
    print(f"⚠️ {DRIVE_PATH} 경로가 없습니다. 폴더를 생성했으니 여기에 파일을 올려주세요.")
```

```python
# 2. 패키지 압축 풀기 및 환경 설정
import zipfile
import shutil

local_zip = os.path.join(DRIVE_PATH, PACKAGE_NAME)
extract_path = '/content/racing_engine'

if os.path.exists(local_zip):
    if os.path.exists(extract_path): shutil.rmtree(extract_path)
    os.makedirs(extract_path)
    
    with zipfile.ZipFile(local_zip, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    
    print(f"✅ 압축 해제 완료: {extract_path}")
    %cd {extract_path}
else:
    print(f"❌ 파일이 없습니다: {local_zip}")
```

```python
# 3. 필수 라이브러리 설치
!pip install google-genai pandas numpy scikit-learn requests beautifulsoup4 aiohttp nest_asyncio
```

```python
# 4. 고성능 AI 가중치 최적화 가동 (Grid Search Deep Mode)
import nest_asyncio
import asyncio
from ml_optimizer import MLOptimizer

# 주피터 노트북 환경에서 비동기 루프를 실행하기 위한 패치
nest_asyncio.apply()

async def run_cloud_opt():
    opt = MLOptimizer()
    # --cloud 플래그 효과를 위해 수동으로 환경 변수 설정 가능
    os.environ['RUN_MODE'] = 'CLOUD'
    await opt.run_optimization()

print("🚀 [Cloud-AI] 대규모 가중치 최적화 시작... (수십 분 소요될 수 있습니다)")
asyncio.run(run_cloud_opt())
```

```python
# 5. 결과 파일 구글 드라이브로 동기화
import shutil
from datetime import datetime

source = '/content/racing_engine/data/optimized_weights.json'
now_str = datetime.now().strftime('%Y%p%d_%H%M')
target = os.path.join(DRIVE_PATH, f'optimized_weights_{now_str}.json')

if os.path.exists(source):
    shutil.copy(source, target)
    print(f"⭐ [완료] 최적 가중치 파일이 드라이브에 저장되었습니다: {target}")
    print("이제 내일 아침 이 파일을 프로젝트의 data/ 폴더에 덮어 쓰시면 됩니다.")
else:
    print("❌ 결과 파일을 찾을 수 없습니다.")
```
